```python
def Edit_Distance_Aligment(seq1, seq2, match, mismatch, gap):

    # # Puntuaciones
    # match = 1
    # mismatch = -1
    # gap = -4
    
    
    #Iniciar Matrices
    score = np.zeros((len(seq1) + 1, len(seq2) + 1))
    backtrack = np.zeros((len(seq1) + 1, len(seq2) + 1), dtype="str")
    
    # Inicializar puntajes
    score[0, :] = np.arange(len(seq2) + 1) * gap
    score[:, 0] = np.arange(len(seq1) + 1) * gap
    
    # Initialar backtrack
    backtrack[0, 1:] = "H"
    backtrack[1:, 0] = "V"
    
    for i in range(1, len(seq1) + 1):
        for j in range(1, len(seq2) + 1):
            # Puntaje diagonal
            if seq1[i - 1] == seq2[j - 1]: 
                scoreD = score[i - 1, j - 1] + match
            else:
                scoreD = score[i - 1, j - 1] + mismatch
    
            # puntaje horizotal
            scoreH = score[i, j - 1] + gap
    
            # puntaje vertical
            scoreV = score[i - 1, j] + gap
    
            # elejir mejor puntaje
            best_score = np.max([scoreD, scoreH, scoreV])
            if scoreD == best_score:
                score[i, j] = scoreD
                backtrack[i, j] = "D"
            elif scoreH == best_score:
                score[i, j] = scoreH
                backtrack[i, j] = "H"
            elif scoreV == best_score:
                score[i, j] = scoreV
                backtrack[i, j] = "V"
    
    # Variables para sequencias alineadas
    aligned_seq1 = ""
    aligned_seq2 = ""
    
    
    i = len(seq1) 
    j = len(seq2)
    
    
    while i != 0 or j != 0:
        # Si es diagonal, simplemente escribe el caracter
        if backtrack[i, j] == "D":
            aligned_seq1 = seq1[i - 1] + aligned_seq1
            aligned_seq2 = seq2[j - 1] + aligned_seq2
            i -= 1
            j -= 1
    
        # Si es horizontal, alinea sequencia 1
        elif backtrack[i, j] == "H":
            aligned_seq1 = "-" + aligned_seq1
            aligned_seq2 = seq2[j - 1] + aligned_seq2
            j -= 1
    
        # Si es vertical, alinea sequencia 2
        elif backtrack[i, j] == "V":
            aligned_seq1 = seq1[i - 1] + aligned_seq1
            aligned_seq2 = "-" + aligned_seq2
            i -= 1     
    
    
    # print(aligned_seq1)
    # print(aligned_seq2)


    # print(score[-1, -1])
    return aligned_seq1, aligned_seq2
```

```python
def CheckDiference(seq1, seq2):
    count = 0
    for i in range(len(seq1)):
        if(seq1[i] != "-" and seq2[i] != "-"):
            if(seq1[i] != seq2[i]):
                count +=1

    return count
```
